```python
# EX-02
# 對影像中的每個通道做運算

import cv2 as cv
import numpy as np

def access_pixels(image):
    print(image.shape)
    height = image.shape[0]
    width = image.shape[1]
    channels = image.shape[2]
    print("width :%s, height :%s, channel :%s"%(width, height, channels))
    for row in range(height):
        for col in range(width):
            for c in range(channels):
                pv = image[row, col, c]
                image[row, col, c] = 255-pv
    cv.imshow("Pixel_demo",image)

print("------------ Hello Python --------------")
src = cv.imread(r'1122_CV_Images/Shohei_2023_WBC_0001.jpg')
cv.namedWindow("Input Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Input Image", src)
access_pixels(src)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-03
# 顯示出演算法運算時間，使用getTickCount()，並用after-before
# getTickFrequency()：用來返回CPU的頻率，單位是：秒。

import cv2 as cv
import numpy as np

def access_pixels(image):
    print(image.shape)
    height = image.shape[0]
    width = image.shape[1]
    channels = image.shape[2]
    print("width :%s, height :%s, channel :%s"%(width, height, channels))
    for row in range(height):
        for col in range(width):
            for c in range(channels):
                pv = image[row, col, c]
                image[row, col, c] = 255-pv
    cv.imshow("Pixel_demo",image)

print("------------ Hello Python --------------")
src = cv.imread(r'1122_CV_Images/Shohei_2023_WBC_0002.jpg')
cv.namedWindow("Input Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Input Image", src)

# 在這邊放入第一個時間點t1
t1 = cv.getTickCount()

# 演算法執行
access_pixels(src)

# 在這邊放入第二個時間點t2
t2 = cv.getTickCount()

# 時間差相減
time = (t2-t1)/cv.getTickFrequency()

# 印出結果
print("time : %s ms"%(time*1000))

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-04
# 新增一 numpy 陣列並填入值(單通道)

import cv2 as cv
import numpy as np

# 創建一個單通道的影像
def Create_Image():
    img = np.zeros([400,400,1],np.uint8)
    img[:,:,0] = np.ones([400,400])*127
    cv.imshow("New Image", img)

Create_Image()
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-05
# 新增一 numpy 陣列並填入值(多通道)

import cv2 as cv
import numpy as np

# 創建一個多通道的影像
def Create_Image():
    img = np.zeros([400,400,3],np.uint8)
    # 讓第一個channel變成255就是藍色
    img[:,:,0] = np.ones([400,400])*255
    cv.imshow("New Image", img)

Create_Image()
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-06
# 利用迴圈加速EX-03

import cv2 as cv
import numpy as np

def Inverse(image):
    dst = cv.bitwise_not(image)
    cv.imshow("pixel_Image", dst)

print("---------------Hello Python--------------------")
src = cv.imread(r'1122_CV_Images/Shohei_2023_WBC_0002.jpg')
cv.namedWindow("Input Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Input Image", src)

# 在這邊放入第一個時間點t1
t1 = cv.getTickCount()

# Inverse演算法執行
Inverse(src)

# 在這邊放入第二個時間點t2
t2 = cv.getTickCount()

# 時間差相減
time = (t2-t1)/cv.getTickFrequency()

# 印出結果
print("time : %s ms"%(time*1000))

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-08
# 影像疊加，指令：cv2.add
# 影像要一樣大，通道數也要一樣，另，如果相加超過255，以255去算
# 以下是範例

import cv2 as cv
import numpy as np

# 定義function add_Demo
def add_Demo(m1,m2):
    dst = cv.add(m1,m2)
    cv.imshow("Add_Demo",dst)

print("---------------Hello Python--------------------")
src1 = cv.imread(r"1122_CV_Images/Background_001.jpg")
src2 = cv.imread(r"1122_CV_Images/Background_002.jpg")
print(src1.shape)
print(src2.shape)

cv.namedWindow("Image1", cv.WINDOW_AUTOSIZE)
cv.imshow("Image1", src1)
cv.namedWindow("Image2", cv.WINDOW_AUTOSIZE)
cv.imshow("Image2", src2)

# 這裡使用function add_Demo
add_Demo(src1, src2)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-09
# 影像疊加，可以搭配權重，可凸顯局部特徵，例如：特定區域變暗、亮
# 公式：dst = img1*α + img2*β + γ 
# α+β不一定要等於1

import cv2 as cv
import numpy as np

print("---------------Hello Python--------------------")
img1 = cv.imread(r"1122_CV_Images/Background_001.jpg")
a = np.ones(img1.shape, dtype=np.uint8)
a *=255

# 搭配權重公式
img2 = cv.addWeighted(img1, 0.8, a, 0.1, 50)

cv.imshow("view1", img1)
cv.imshow("view2", img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-10
# 重複曝光技術：結合了兩張或更多張圖像，創造出一個疊加的視覺效果。
# 應用：藝術和設計、廣告和市場營銷、影像合成

import cv2 as cv
import numpy as np
src1 = cv.imread(r"1122_CV_Images/Background_001.jpg")
src2 = cv.imread(r"1122_CV_Images/Background_002.jpg")

# 給定α、β、γ的值
alpha = 0.7
beta = 0.4
gamma = 1

dst = cv.addWeighted(src1, alpha, src2, beta, gamma)

cv.imshow("src1 + src2", dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-11
# 影像相減，指令：cv2.substract
# 影像要一樣大，通道數也要一樣，另，如果相加超過0，以0去算
# 以下是範例

import cv2 as cv
import numpy as np

# 定義function subtract_Demo
def subtract_Demo(m1,m2):
    dst = cv.subtract(m1,m2)
    cv.imshow("Subtract_Demo",dst)

print("---------------Hello Python--------------------")
src1 = cv.imread(r"1122_CV_Images/Background_001.jpg")
src2 = cv.imread(r"1122_CV_Images/Background_002.jpg")
print(src1.shape)
print(src2.shape)

cv.namedWindow("Image1", cv.WINDOW_AUTOSIZE)
cv.imshow("Image1", src1)
cv.namedWindow("Image2", cv.WINDOW_AUTOSIZE)
cv.imshow("Image2", src2)

# 這裡使用function subtract_Demo
subtract_Demo(src1, src2)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-12
# 影像位元運算系列一、邏輯And運算，指令：cv2.bitwise_and():

import cv2 as cv
import numpy as np

print("---------------Hello Python--------------------")
src1 = cv.imread(r"1122_CV_Images/Background_001.jpg", cv.IMREAD_GRAYSCALE)
src2 = np.zeros(src1.shape,dtype=np.uint8)

# 設定遮罩高與寬
src2[60:420, 120:370] = 255
dst = cv.bitwise_and(src1,src2)

# 顯示影像
cv.imshow("Source Image", src1)
cv.imshow("Mask Image", src2)
cv.imshow("Result Image", dst)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-13
# 影像位元運算系列二、邏輯or運算，指令：cv2.bitwise_or():

import cv2 as cv
import numpy as np

print("---------------Hello Python--------------------")
src1 = cv.imread(r"1122_CV_Images/Background_001.jpg", cv.IMREAD_GRAYSCALE)
src2 = np.zeros(src1.shape,dtype=np.uint8)

# 設定遮罩高與寬
src2[60:420, 120:370] = 255

# 這邊改成cv.bitwise_or
dst = cv.bitwise_or(src1,src2)

# 顯示影像
cv.imshow("Source Image", src1)
cv.imshow("Mask Image", src2)
cv.imshow("Result Image", dst)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-14
# 影像位元運算系列三、邏輯not運算，指令：cv2.bitwise_not():

import cv2 as cv
import numpy as np

# 讀取檔案&執行not運算
src = cv.imread(r"1122_CV_Images/Julia_001.jpg")
dst = cv.bitwise_not(src)
dst2 = cv.bitwise_not(dst)

# 顯示影像
cv.imshow("Julia", src)
cv.imshow("Not Julia", dst)
cv.imshow("Not Julia Not", dst2)

# 刪除視窗
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# EX-15
# 影像位元運算系列四、邏輯xor運算，指令：cv2.bitwise_xor():
# 定義：兩個位不相同的情況下返回真（1），相同則返回假（0）
# 使用時機1：在監控攝像中，XOR可以用來檢測畫面中移動的物體。
# 使用時機2：在數位圖像隱寫術中，XOR運算可以用來隱藏一張圖像（信息）在另一張圖像中。

import cv2 as cv
import numpy as np

# 讀取檔案&執行not運算
src1 = cv.imread(r"1122_CV_Images/Julia_001.jpg")
src2 = np.zeros(src1.shape,dtype=np.uint8)

# 建立mask白色區域
src2[:,120:360,:] = 255

# 執行xor
dst = cv.bitwise_xor(src1, src2)

# 顯示影像
cv.imshow("Julia", src1)
cv.imshow("Mask", src2)
cv.imshow("Julia xor operation", dst)

# 刪除視窗
cv.waitKey(0)
cv.destroyAllWindows()
```
