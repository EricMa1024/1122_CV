```python
# 展示一個1D的convolution

import numpy as np
x = np.array([1,2,4,3,2,1,1])
h = np.array([1,2,3,1,1])

# 在做convolve時，h array會反過來
y = np.convolve(x,h,'full')
print('Convolution y = ', y)
```

```python
# 平均濾波(Averaging)的範例
# 函數：blur

import cv2 as cv
import numpy as np

def blur_demo(image):
    dst = cv.blur(image,(5,5))
    cv.imshow("blur",dst)

print("--------- Hello Python -------------")

# 讀取雜訊影像碩專/1122/1122_CV/1122_CV_Images/Julia_001.jpg
# src = cv.imread("Lenanooise.jpg")
src = cv.imread(r"1122_CV_Images/Julia_001.jpg")

# 自動調整視窗大小
cv.namedWindow("Input Image", cv.WINDOW_AUTOSIZE)

# 顯示圖片
cv.imshow("Input Image", src)
blur_demo(src)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 使用方框濾波器:boxFilter()
# 函數：boxfilter

import cv2 as cv
import numpy as np

# -1：跟原始的圖同樣的深度
# normalize=True：均值濾波；False：只加總部平均，超過255就以255算
def boxFilter_demo(image):
    dst = cv.boxFilter(image,-1,(3,3),normalize=False)
    cv.imshow("boxFilter",dst)

print("--------- Hello Python -------------")

# 碩專/1122/1122_CV/1122_CV_Images/Julia_001.jpg

src = cv.imread(r"1122_CV_Images/Julia_001.jpg")

# 自動調整視窗大小
cv.namedWindow("Input Image", cv.WINDOW_AUTOSIZE)

# 顯示圖片
cv.imshow("Input Image", src)
boxFilter_demo(src)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 中值濾波(Median Filter)的範例
# 函數：medianBlur()

import cv2 as cv
import numpy as np

# 生成一個矩陣，範例是3z3，裡面的值都給100
src = np.ones((3,3), np.float32)*100

# 在矩陣的[1,1]位置把值15放進去
src[1,1] = 15
print(f"src = \n{src}")

# 利用平均濾波把15給濾掉
dst = cv.medianBlur(src,3)
print(f"dst = \n{dst}")
```

```python
# 使用中值濾波器:medianBlur()
# 用途：把雜質給濾掉

import cv2 as cv
import numpy as np


def MedianBlurFilter_demo(image):
    dst = cv.medianBlur(image,5)
    cv.imshow("medianBlur",dst)

print("--------- Hello Python -------------")

src = cv.imread(r"1122_CV_Images/image/Lenanoise.jpg")

# 自動調整視窗大小
cv.namedWindow("Input Image", cv.WINDOW_AUTOSIZE)

# 顯示圖片
cv.imshow("Input Image", src)
MedianBlurFilter_demo(src)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 取得高斯濾波器:GaussianKernel()
# 利用自己 乘上 自己的Transpose

import cv2 as cv
import numpy as np

kernel_size = 3
sigma = 0.85

# 得到1-D Gaussian kernel
kernel_1d = cv.getGaussianKernel(kernel_size, sigma)
print(kernel_1d)

# 生成1D Gaussian kernel 後乘以自己的轉置得到2D的Gaussian Kernel
kernel_2d = kernel_1d * kernel_1d.T
print(kernel_2d)
```

```python
# 使用高斯濾波器:GaussianBlur()

import cv2 as cv

# 碩專/1122/1122_CV/1122_CV_Images/Julia_002.jpg
src = cv.imread(r"1122_CV_Images/Julia_002.jpg")

# 改kernel，從3x3-> 5x5-> 15x15
dst1 = cv.GaussianBlur(src,(3,3),0,0)
dst2 = cv.GaussianBlur(src,(5,5),0,0)
dst3 = cv.GaussianBlur(src,(15,15),0,0)

# 顯示圖片
cv.imshow("src", src)
cv.imshow("3x3 kernel", dst1)
cv.imshow("5x5 kernel", dst2)
cv.imshow("15x15 kernel", dst3)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 使用高斯雙邊濾波器:bilateralFilter()
# 優點是：去噪保邊
# 缺點是：運算量大

"""
    老師提供的註解如下：
    高斯雙邊模糊
    d:即 distance，常規為0，像素的臨域直徑，可由sigmaColor和sigmaSpace
    sigmaColor：盡量取大，目的將小的差異模糊掉，噪聲去掉
    sigmaSpace：盡量取小，這樣核就會小點，主要的差異就會留下來
"""

import cv2 as cv

def bi_demo(image):
    dst = cv.bilateralFilter(image,0,100,15)
    cv.imshow("bi_demo",dst)

src = cv.imread(r"1122_CV_Images/Julia_002.jpg")

# 自動調整視窗大小
cv.namedWindow("Input Image", cv.WINDOW_AUTOSIZE)

# 顯示圖片
cv.imshow("Input Image", src)
bi_demo(src)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 使用自己定義的濾波器

import cv2 as cv
import numpy as np

src = cv.imread(r"1122_CV_Images/Julia_002.jpg")
kernel = np.ones((9,9),np.float32)/81

dst = cv.filter2D(src,-1,kernel)

# 顯示圖片
cv.imshow("src Image", src)
cv.imshow("dst Image", dst)

cv.waitKey(0)
cv.destroyAllWindows()
```
