```python
# 讀取圖檔
# 係數給定0代表讀出"灰階格式"
img = cv2.imread(r"1122_CV_Images/Shohei_2023_WBC_0001.jpg")
```

```python
# 顯示圖片
cv2.imshow('Shohei Ohtani in 2023',img) #不能單獨使用，imshow會這樣，但是matplotlib不會

# 按下任意鍵則關閉所有視窗
cv2.waitKey(0)

# destroyAllWindows() 銷毀所有窗口
cv2.destroyAllWindows()
```

```python
# 注意：opencv的格式是BGR，處理彩色影像時要轉換成RGB，語法是：cv2.COLOR_BGR2RGB
import cv2
from matplotlib import pyplot as plt

# 讀取圖檔
img = cv2.imread(r"1122_CV_Images/Shohei_2023_WBC_0002.jpg")

# 查看資料型態
type(img)

# 顯示影像格式
img.shape

# 顯示圖片
# 注意：opencv的格式是BGR，處理彩色影像時要轉換成RGB，語法是：cv2.COLOR_BGR2RGB
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Shohei Ohtani in 2023 WBC')
plt.show()
cv2.waitKey(0)
```

```python
# 影像儲存，參數名稱：cv2.imwrite()
import cv2

# imread(path,0) 的 0 代表以灰階來讀取，並且轉換成灰階並儲存
img=cv2.imread(r"1122_CV_Images/Shohei_2023_WBC_0002.jpg",0)
cv2.imwrite("1122_CV_Images/Shohei_2023_WBC_0002_GrayScale.jpg",img)
```
